```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
url = 'amz_uk_price_prediction_dataset.csv'
df = pd.read_csv(url)
```

### Part 1: Understanding Product Categories

1. **Frequency Tables**:

```python
df['category'].value_counts()
```

```python
frequency_table = df['category'].value_counts().head(5)
frequency_table
```

2. **Visualizations**:

```python
plt.figure(figsize=(13, 4))
sns.barplot(x=frequency_table.index, y=frequency_table.values, palette='pastel', hue=frequency_table.index)
plt.ylabel('Number of products')
plt.title('Distribution of products')
```

```python
plt.figure(figsize=(10, 10))
frequency_table.plot.pie(autopct='%1.1f%%', startangle=25, colors=sns.color_palette("pastel"))
```

### Part 2: Delving into Product Pricing

2. **Measures of Dispersion**

```python
print('Mean: ', df['price'].mean())
print('Median: ', df['price'].median())
print('Mode: ', df['price'].mode()[0])
```

2. **Measures of Dispersion**

```python
print('Variance: ', df['price'].var())
print('Standard deviation: ', df['price'].std())
print('Range: ', df['price'].max(), '-', df['price'].min(),':',df['price'].max()-df['price'].min())
print('Quartiles:\n', df['price'].quantile([0.25, 0.5, 0.75]))
```

3. **Visualizations**

```python
df_vis = df[df['price']<300]
```

```python
sns.histplot(df_vis['price'], kde=True, bins=30, color='skyblue')
plt.title('Distribución de los Precios de los Productos')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
```

```python
sns.boxplot(data=df_vis['price'], color='skyblue')
```

### Part 3: Unpacking Product Ratings

1. **Measures of Centrality**

```python
print('Mean: ', df['stars'].mean())
print('Median: ', df['stars'].median())
print('Mode: ', df['stars'].mode()[0])
```

2. **Measures of Dispersion**

```python
print('Variance: ', df['stars'].var())
print('Standard deviation: ', df['stars'].std())
print('Range: ', df['stars'].max(), '-', df['stars'].min(),':',df['stars'].max()-df['stars'].min())
print('Quartiles:\n', df['stars'].quantile([0.25, 0.5, 0.75]))
```

3. **Shape of the Distribution**

```python
df['stars'].skew()
```

```python
df['stars'].kurtosis()
```

4. **Visualizations**

```python
sns.kdeplot(df['stars'], color='skyblue', shade=True)
plt.title('Distribución de los Precios de las stars')
plt.xlabel('Stars')
plt.ylabel('Frecuencia')
```
